A simple function to plot distribution

```python
# imports
import seaborn as sns
import matplotlib.pyplot as plt

# Configure visualisations
%matplotlib inline
sns.set_style( 'white' )
```

Function takes 3 main arguments
- df: dataframe to visualise
- var: x axis values
- target: y axis values

row and col can be passed in to split data across multiple categories

```python
def plot_distribution( df , var , target , **kwargs ):
    row = kwargs.get( 'row' , None )
    col = kwargs.get( 'col' , None )
    facet = sns.FacetGrid( df , hue=target , aspect=4 , row = row , col = col )
    facet.map( sns.kdeplot , var , shade= True )
    facet.set( xlim=( 0 , df[ var ].max() ) )
    facet.add_legend()
```

Sample:

```python
import pandas as pd
import numpy as np

y = (pd.Series(np.random.randn(1000)) > 0.5).astype(int)
x = pd.Series(np.random.randn(1000)) * 100

df = pd.DataFrame({'x': x, 'y': y})
plot_distribution( df , var = 'x' , target = 'y' )
```
